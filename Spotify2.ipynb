```python
from concurrent.futures import ThreadPoolExecutor

def search_spotify(track_name, spotify_obj):
    return spotify_obj.search(q='track:' + track_name, type='track')

spotify = spotipy.Spotify()
work_function = lambda x: search_spotify(x, spotify)
output=[]

def multithread(work_list):
    with ThreadPoolExecutor(max_workers=5) as executor:
        for work in work_list:
            future = executor.submit(work_function, work)
    return future.result()
# 5 workers @ 43 secs

def multithread2(work_list):
    output = []
    with ThreadPoolExecutor(max_workers=5) as executor:
        for work in work_list:
            future = executor.submit(work_function, work)
            output.append(future)
    return map(lambda x: x.result(), output)

def multithread3(work_list):
    with ThreadPoolExecutor(max_workers=5) as executor:
        output = executor.map(work_function, work_list)    
    return output
```

```python
def query_all_possibilities(input_str):
    output = []
    spotify = spotipy.Spotify()
    for n in range(len(input_str.split()))[::-1]:
        ngram_set = find_n_grams(input_str, n + 1) 
        for ngram in ngram_set:
            results = search_spotify(ngram, spotify)
            if results: 
                validated_results = search_through_results(ngram, results)
                if validated_results: 
                    output.append((ngram, n + 1))
    return output
```
